```python
import numpy as np
import random
from typing import List, Tuple, Dict, Set, Optional, Any
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import copy
from collections import deque, defaultdict
import warnings
import pandas as pd
from scipy import stats
import time

warnings.filterwarnings('ignore')

np.random.seed(42)
torch.manual_seed(42)
random.seed(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ============================================================================
# 1. IMPROVED GAME ENVIRONMENT
# ============================================================================

class MinesweeperGame:
    def __init__(self, height=22, width=22, num_mines=80):
        self.height = height
        self.width = width
        self.num_mines = num_mines
        self.reset()
    
    def reset(self, first_click=None):
        self.board = np.zeros((self.height, self.width), dtype=int)
        self.revealed = np.zeros((self.height, self.width), dtype=bool)
        self.flagged = np.zeros((self.height, self.width), dtype=bool)
        self.game_over = False
        self.won = False
        self.mines_triggered = 0
        
        all_cells = [(r, c) for r in range(self.height) for c in range(self.width)]
        
        if first_click is not None:
            # Ensure first click is safe
            safe_radius = [(dr, dc) for dr in [-1, 0, 1] for dc in [-1, 0, 1]]
            safe_cells = []
            for dr, dc in safe_radius:
                r, c = first_click[0] + dr, first_click[1] + dc
                if 0 <= r < self.height and 0 <= c < self.width:
                    safe_cells.append((r, c))
            safe_cells = list(set(safe_cells))
            mine_cells = [cell for cell in all_cells if cell not in safe_cells]
            mine_positions = random.sample(mine_cells, min(self.num_mines, len(mine_cells)))
        else:
            mine_positions = random.sample(all_cells, self.num_mines)
        
        for r, c in mine_positions:
            self.board[r, c] = 1
        
        self.clues = np.zeros((self.height, self.width), dtype=int)
        for r in range(self.height):
            for c in range(self.width):
                if self.board[r, c] == 0:
                    self.clues[r, c] = self._count_mines(r, c)
        
        return self.get_state()
    
    def _get_neighbors(self, r, c):
        neighbors = []
        for dr in [-1, 0, 1]:
            for dc in [-1, 0, 1]:
                if dr == 0 and dc == 0:
                    continue
                nr, nc = r + dr, c + dc
                if 0 <= nr < self.height and 0 <= nc < self.width:
                    neighbors.append((nr, nc))
        return neighbors
    
    def _count_mines(self, r, c):
        count = 0
        for nr, nc in self._get_neighbors(r, c):
            if self.board[nr, nc] == 1:
                count += 1
        return count
    
    def reveal(self, r, c):
        if self.game_over or self.revealed[r, c] or self.flagged[r, c]:
            return None
        
        self.revealed[r, c] = True
        
        if self.board[r, c] == 1:
            self.mines_triggered += 1
            self.game_over = True
            self.won = False
            return -1
        
        if self.clues[r, c] == 0:
            self._flood_fill(r, c)
        
        if np.sum(self.revealed) == self.height * self.width - self.num_mines:
            self.game_over = True
            self.won = True
        
        return self.clues[r, c]
    
    def _flood_fill(self, r, c):
        stack = [(r, c)]
        visited = set()
        
        while stack:
            cr, cc = stack.pop()
            if (cr, cc) in visited:
                continue
            visited.add((cr, cc))
            
            for nr, nc in self._get_neighbors(cr, cc):
                if not self.revealed[nr, nc] and not self.flagged[nr, nc]:
                    self.revealed[nr, nc] = True
                    if self.clues[nr, nc] == 0:
                        stack.append((nr, nc))
    
    def get_state(self):
        state = np.zeros((5, self.height, self.width), dtype=np.float32)
        
        # Channel 0: Revealed clues (normalized 0-1)
        for r in range(self.height):
            for c in range(self.width):
                if self.revealed[r, c]:
                    state[0, r, c] = self.clues[r, c] / 8.0
                else:
                    state[0, r, c] = -1.0
        
        # Channel 1: Mine probability estimate
        total_hidden = np.sum(~self.revealed & ~self.flagged)
        if total_hidden > 0:
            remaining_mines = self.num_mines - np.sum(self.flagged)
            prob = remaining_mines / max(1, total_hidden)
            state[1, :, :] = prob
        
        # Channel 2: Flag positions
        state[2, self.flagged] = 1.0
        
        # Channel 3: Revealed cells mask
        state[3, self.revealed] = 1.0
        
        # Channel 4: Border distance (helps with edge patterns)
        for r in range(self.height):
            for c in range(self.width):
                dist_to_edge = min(r, self.height-1-r, c, self.width-1-c)
                state[4, r, c] = dist_to_edge / (self.height // 2)
        
        return state
    
    def get_ground_truth(self):
        gt = self.board.copy().astype(np.float32)
        gt[self.revealed] = -1
        return gt
    
    def get_available_moves(self):
        moves = []
        for r in range(self.height):
            for c in range(self.width):
                if not self.revealed[r, c] and not self.flagged[r, c]:
                    moves.append((r, c))
        return moves

# ============================================================================
# 2. IMPROVED LOGIC BOT
# ============================================================================

class LogicBot:
    def __init__(self, game):
        self.game = game
        self.cells_remaining = set((r, c) for r in range(game.height) for c in range(game.width))
        self.inferred_safe = set()
        self.inferred_mine = set()
        self.clue_numbers = {}
        self.moves_made = 0
    
    def make_move(self, allow_guess=True):
        self.moves_made += 1
        
        # Check if we have inferred safe cells
        if self.inferred_safe:
            r, c = self.inferred_safe.pop()
            result = self.game.reveal(r, c)
            if result is not None:
                if result == -1:
                    return False, 'mine'
                else:
                    self.clue_numbers[(r, c)] = result
            self.cells_remaining.discard((r, c))
            self._update_inferences()
            return True, 'inference'
        
        # If no inferences, make a random move
        if not self.cells_remaining:
            return False, 'no_moves'
        
        if allow_guess:
            r, c = random.choice(list(self.cells_remaining))
            result = self.game.reveal(r, c)
            if result is not None:
                if result == -1:
                    return False, 'mine'
                else:
                    self.clue_numbers[(r, c)] = result
            self.cells_remaining.discard((r, c))
            self._update_inferences()
            return True, 'guess'
        
        return False, 'stuck'
    
    def _update_inferences(self):
        changed = True
        while changed:
            changed = False
            
            clue_cells = [cell for cell in self.clue_numbers.keys() 
                         if self.game.revealed[cell]]
            
            for r, c in clue_cells:
                clue = self.clue_numbers[(r, c)]
                neighbors = self.game._get_neighbors(r, c)
                
                # Count inferred mines among neighbors
                inferred_mines = sum(1 for nr, nc in neighbors 
                                   if (nr, nc) in self.inferred_mine)
                
                # Count revealed safe neighbors
                safe_neighbors = sum(1 for nr, nc in neighbors 
                                   if self.game.revealed[nr, nc] or (nr, nc) in self.inferred_safe)
                
                # Get unrevealed neighbors not yet inferred
                unrevealed_neighbors = [(nr, nc) for nr, nc in neighbors 
                                      if not self.game.revealed[nr, nc] 
                                      and (nr, nc) not in self.inferred_mine 
                                      and (nr, nc) not in self.inferred_safe]
                
                # Rule 1: If clue - inferred_mines == # unrevealed, all unrevealed are mines
                if clue - inferred_mines == len(unrevealed_neighbors) and len(unrevealed_neighbors) > 0:
                    for nr, nc in unrevealed_neighbors:
                        if (nr, nc) not in self.inferred_mine:
                            self.inferred_mine.add((nr, nc))
                            self.cells_remaining.discard((nr, nc))
                            changed = True
                
                # Rule 2: If total_neighbors - clue - safe_neighbors == # unrevealed, all unrevealed are safe
                total_neighbors = len(neighbors)
                if (total_neighbors - clue) - safe_neighbors == len(unrevealed_neighbors) and len(unrevealed_neighbors) > 0:
                    for nr, nc in unrevealed_neighbors:
                        if (nr, nc) not in self.inferred_safe:
                            self.inferred_safe.add((nr, nc))
                            self.cells_remaining.discard((nr, nc))
                            changed = True
    
    def run_game(self, max_steps=1000):
        steps = 0
        while not self.game.game_over and steps < max_steps:
            success, move_type = self.make_move()
            if not success:
                break
            steps += 1
        
        return {
            'won': self.game.won,
            'steps': steps,
            'mines_triggered': self.game.mines_triggered,
            'moves_made': self.moves_made
        }

# ============================================================================
# 3. ENHANCED DATASET GENERATION
# ============================================================================

class MinesweeperDataset(Dataset):
    def __init__(self, num_samples=10000, height=22, width=22, 
                 num_mines=80, difficulty='medium', include_logic_data=True):
        self.height = height
        self.width = width
        self.num_mines = num_mines
        self.num_samples = num_samples
        self.difficulty = difficulty
        self.include_logic_data = include_logic_data
        
        self.states = []
        self.mine_labels = []
        self.survival_labels = []
        self.move_labels = []
        
        print(f"Generating {num_samples} training samples ({difficulty})...")
        self._generate_data()
    
    def _generate_data(self):
        for _ in tqdm(range(self.num_samples)):
            game = MinesweeperGame(self.height, self.width, self.num_mines)
            
            # Start with first safe click
            first_r, first_c = np.random.randint(0, self.height), np.random.randint(0, self.width)
            game.reset(first_click=(first_r, first_c))
            game.reveal(first_r, first_c)
            
            # Play some moves to create partial board
            steps = np.random.randint(3, 15)
            bot = LogicBot(game)
            
            survival_count = 0
            for step in range(steps):
                success, move_type = bot.make_move()
                if not success:
                    break
                survival_count += 1
            
            # Only save if we have enough revealed cells
            if np.sum(game.revealed) > 5 and np.sum(game.revealed) < (self.height * self.width - self.num_mines - 5):
                state = game.get_state()
                mine_label = game.get_ground_truth()
                
                # Calculate survival duration if logic bot continues
                if self.include_logic_data:
                    test_game = copy.deepcopy(game)
                    test_bot = LogicBot(test_game)
                    remaining_survival = 0
                    while not test_game.game_over and remaining_survival < 50:
                        success, _ = test_bot.make_move()
                        if not success:
                            break
                        remaining_survival += 1
                    
                    # Create move preference labels
                    move_prefs = np.zeros((self.height, self.width), dtype=np.float32)
                    available_moves = game.get_available_moves()
                    if available_moves:
                        # Prefer moves near revealed cells with clues
                        for r, c in available_moves:
                            neighbors = game._get_neighbors(r, c)
                            clue_sum = 0
                            revealed_neighbors = 0
                            for nr, nc in neighbors:
                                if game.revealed[nr, nc]:
                                    revealed_neighbors += 1
                                    clue_sum += game.clues[nr, nc]
                            if revealed_neighbors > 0:
                                move_prefs[r, c] = clue_sum / (revealed_neighbors * 8.0)
                    
                    self.survival_labels.append(remaining_survival)
                    self.move_labels.append(move_prefs)
                
                self.states.append(state)
                self.mine_labels.append(mine_label)
    
    def __len__(self):
        return len(self.states)
    
    def __getitem__(self, idx):
        state = torch.FloatTensor(self.states[idx])
        mine_label = torch.FloatTensor(self.mine_labels[idx])
        
        if self.include_logic_data and len(self.survival_labels) > idx:
            survival = torch.FloatTensor([self.survival_labels[idx] / 50.0])  # Normalized
            move_pref = torch.FloatTensor(self.move_labels[idx])
            return state, mine_label, survival, move_pref
        else:
            return state, mine_label

# ============================================================================
# 4. NEURAL NETWORK ARCHITECTURES
# ============================================================================

class AttentionBlock(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.query = nn.Conv2d(channels, channels // 8, 1)
        self.key = nn.Conv2d(channels, channels // 8, 1)
        self.value = nn.Conv2d(channels, channels, 1)
        self.gamma = nn.Parameter(torch.zeros(1))
        self.softmax = nn.Softmax(dim=-1)
    
    def forward(self, x):
        batch_size, C, H, W = x.shape
        
        query = self.query(x).view(batch_size, -1, H * W).permute(0, 2, 1)
        key = self.key(x).view(batch_size, -1, H * W)
        value = self.value(x).view(batch_size, -1, H * W)
        
        attention = torch.bmm(query, key) / (C ** 0.5)
        attention = self.softmax(attention)
        
        out = torch.bmm(value, attention.permute(0, 2, 1))
        out = out.view(batch_size, C, H, W)
        
        return self.gamma * out + x

class ResBlock(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.conv1 = nn.Conv2d(channels, channels, 3, padding=1)
        self.bn1 = nn.BatchNorm2d(channels)
        self.conv2 = nn.Conv2d(channels, channels, 3, padding=1)
        self.bn2 = nn.BatchNorm2d(channels)
        self.relu = nn.ReLU(inplace=True)
    
    def forward(self, x):
        residual = x
        out = self.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        out += residual
        out = self.relu(out)
        return out

# ============================================================================
# TASK 1: MINE PREDICTOR
# ============================================================================

class MinePredictor(nn.Module):
    def __init__(self, input_channels=5, hidden_channels=128):
        super().__init__()
        
        # Initial convolution
        self.conv1 = nn.Conv2d(input_channels, hidden_channels, 3, padding=1)
        self.bn1 = nn.BatchNorm2d(hidden_channels)
        
        # Residual blocks with attention
        self.res_blocks = nn.ModuleList([
            ResBlock(hidden_channels) for _ in range(6)
        ])
        
        # Attention blocks
        self.attention_blocks = nn.ModuleList([
            AttentionBlock(hidden_channels) for _ in range(2)
        ])
        
        # Multi-scale feature aggregation
        self.conv2 = nn.Conv2d(hidden_channels, hidden_channels // 2, 3, padding=1)
        self.bn2 = nn.BatchNorm2d(hidden_channels // 2)
        
        self.conv3 = nn.Conv2d(hidden_channels // 2, hidden_channels // 4, 3, padding=1)
        self.bn3 = nn.BatchNorm2d(hidden_channels // 4)
        
        # Final prediction
        self.final_conv = nn.Conv2d(hidden_channels // 4, 1, 1)
        
        # Global context
        self.global_pool = nn.AdaptiveAvgPool2d(1)
        self.global_fc = nn.Linear(hidden_channels, hidden_channels // 4)
        
    def forward(self, x):
        # Initial feature extraction
        x = F.relu(self.bn1(self.conv1(x)))
        
        # Residual processing
        for res_block in self.res_blocks:
            x = res_block(x)
        
        # Attention for long-range dependencies
        for att_block in self.attention_blocks:
            x = att_block(x)
        
        # Global context
        global_feat = self.global_pool(x).squeeze(-1).squeeze(-1)
        global_feat = self.global_fc(global_feat).unsqueeze(-1).unsqueeze(-1)
        
        # Multi-scale features
        x = F.relu(self.bn2(self.conv2(x)))
        x = F.relu(self.bn3(self.conv3(x)))
        
        # Add global context
        x = x + global_feat
        
        # Final prediction
        x = self.final_conv(x)
        x = torch.sigmoid(x)
        
        return x

# ============================================================================
# TASK 2: ACTOR-CRITIC NETWORK
# ============================================================================

class ActorCriticNetwork(nn.Module):
    def __init__(self, input_channels=5, hidden_channels=128):
        super().__init__()
        
        # Shared encoder
        self.shared_encoder = nn.Sequential(
            nn.Conv2d(input_channels, hidden_channels, 3, padding=1),
            nn.BatchNorm2d(hidden_channels),
            nn.ReLU(inplace=True),
            ResBlock(hidden_channels),
            ResBlock(hidden_channels),
            AttentionBlock(hidden_channels),
        )
        
        # Critic head: predicts survival duration
        self.critic_head = nn.Sequential(
            nn.Conv2d(hidden_channels, hidden_channels // 2, 3, padding=1),
            nn.BatchNorm2d(hidden_channels // 2),
            nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(hidden_channels // 2, 64),
            nn.ReLU(inplace=True),
            nn.Linear(64, 1),
            nn.Sigmoid()  # Output normalized survival probability
        )
        
        # Actor head: predicts move values
        self.actor_head = nn.Sequential(
            nn.Conv2d(hidden_channels, hidden_channels // 2, 3, padding=1),
            nn.BatchNorm2d(hidden_channels // 2),
            nn.ReLU(inplace=True),
            nn.Conv2d(hidden_channels // 2, 1, 1),  # Single channel output
        )
    
    def forward(self, x):
        shared = self.shared_encoder(x)
        
        # Critic: overall state value
        critic_out = self.critic_head(shared)
        
        # Actor: move values
        actor_out = self.actor_head(shared).squeeze(1)  # Shape: (batch, H, W)
        
        # Apply mask: unrevealed cells only
        mask = (x[:, 3, :, :] == 0).float()  # Channel 3 is revealed mask
        actor_out = actor_out * mask
        
        # Softmax over all cells
        batch_size, H, W = actor_out.shape
        actor_probs = F.softmax(actor_out.view(batch_size, -1), dim=-1)
        actor_probs = actor_probs.view(batch_size, H, W)
        
        return critic_out, actor_probs

# ============================================================================
# TASK 3: SEQUENTIAL THINKER
# ============================================================================

class SequentialThinker(nn.Module):
    def __init__(self, input_channels=5, hidden_channels=128, num_steps=10):
        super().__init__()
        self.num_steps = num_steps
        
        # Initial feature extraction
        self.init_conv = nn.Conv2d(input_channels, hidden_channels, 3, padding=1)
        self.init_bn = nn.BatchNorm2d(hidden_channels)
        
        # Sequential reasoning blocks
        self.reasoning_cells = nn.ModuleList([
            nn.GRUCell(hidden_channels, hidden_channels) for _ in range(num_steps)
        ])
        
        # Convolutional processors for each step
        self.conv_processors = nn.ModuleList([
            nn.Sequential(
                nn.Conv2d(hidden_channels, hidden_channels, 3, padding=1),
                nn.BatchNorm2d(hidden_channels),
                nn.ReLU(inplace=True),
                nn.Conv2d(hidden_channels, hidden_channels, 3, padding=1),
                nn.BatchNorm2d(hidden_channels),
            ) for _ in range(num_steps)
        ])
        
        # Prediction heads for each step
        self.prediction_heads = nn.ModuleList([
            nn.Sequential(
                nn.Conv2d(hidden_channels, hidden_channels // 2, 3, padding=1),
                nn.BatchNorm2d(hidden_channels // 2),
                nn.ReLU(inplace=True),
                nn.Conv2d(hidden_channels // 2, 1, 1),
                nn.Sigmoid()
            ) for _ in range(num_steps)
        ])
        
    def forward(self, x, steps=None, return_all=False):
        if steps is None:
            steps = self.num_steps
        
        # Initial features
        x = F.relu(self.init_bn(self.init_conv(x)))
        batch_size, C, H, W = x.shape
        
        # Initialize hidden state
        hidden = x.clone()
        
        intermediate_preds = []
        intermediate_features = []
        
        # Sequential reasoning
        for i in range(steps):
            # Process through GRU cell (flatten spatial dims)
            hidden_flat = hidden.view(batch_size, C, -1).permute(0, 2, 1)  # (B, H*W, C)
            x_flat = x.view(batch_size, C, -1).permute(0, 2, 1)
            
            # Apply GRU cell to each spatial location
            new_hidden = []
            for loc in range(H * W):
                h_loc = self.reasoning_cells[i](x_flat[:, loc, :], hidden_flat[:, loc, :])
                new_hidden.append(h_loc)
            
            hidden = torch.stack(new_hidden, dim=1).permute(0, 2, 1).view(batch_size, C, H, W)
            
            # Convolutional processing
            conv_out = self.conv_processors[i](hidden)
            hidden = F.relu(hidden + conv_out)
            
            # Make prediction at this step
            pred = self.prediction_heads[i](hidden)
            intermediate_preds.append(pred)
            intermediate_features.append(hidden.clone())
        
        if return_all:
            return intermediate_preds, intermediate_features
        else:
            return intermediate_preds[-1]

# ============================================================================
# LOSS FUNCTIONS
# ============================================================================

class WeightedBCELoss(nn.Module):
    def __init__(self, pos_weight=1.0, neg_weight=0.5):
        super().__init__()
        self.pos_weight = pos_weight
        self.neg_weight = neg_weight
    
    def forward(self, pred, target):
        # pred: (B, 1, H, W), target: (B, H, W) with -1 for ignored cells
        
        # Create mask for valid cells (not revealed)
        mask = (target != -1).float().unsqueeze(1)  # (B, 1, H, W)
        
        # Convert target to binary (0 for safe, 1 for mine)
        target_binary = torch.clamp(target, 0, 1).unsqueeze(1)  # (B, 1, H, W)
        
        # Calculate sample weights
        weights = torch.ones_like(pred)
        weights[target_binary == 1] = self.pos_weight
        weights[target_binary == 0] = self.neg_weight
        
        # Apply mask
        weights = weights * mask
        
        # Calculate weighted BCE
        pred = torch.clamp(pred, 1e-7, 1 - 1e-7)
        bce = - (target_binary * torch.log(pred) + (1 - target_binary) * torch.log(1 - pred))
        weighted_bce = bce * weights
        
        # Only average over valid cells
        valid_count = torch.max(torch.tensor(1.0).to(pred.device), mask.sum())
        loss = weighted_bce.sum() / valid_count
        
        return loss

class SurvivalLoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.mse = nn.MSELoss()
    
    def forward(self, pred, target):
        return self.mse(pred.squeeze(), target)

# ============================================================================
# TRAINING FUNCTIONS
# ============================================================================

def train_mine_predictor(model, train_loader, val_loader, num_epochs=20, lr=0.001):
    model = model.to(device)
    criterion = WeightedBCELoss(pos_weight=2.0, neg_weight=1.0)
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=3, factor=0.5)
    
    train_losses = []
    val_losses = []
    val_accuracies = []
    
    print("Training Mine Predictor...")
    for epoch in range(num_epochs):
        # Training phase
        model.train()
        epoch_train_loss = 0
        batch_count = 0
        
        for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Train"):
            if len(batch) == 4:
                states, mine_labels, _, _ = batch
            else:
                states, mine_labels = batch
            
            states = states.to(device)
            mine_labels = mine_labels.to(device)
            
            optimizer.zero_grad()
            predictions = model(states)
            loss = criterion(predictions, mine_labels)
            loss.backward()
            
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            
            epoch_train_loss += loss.item()
            batch_count += 1
        
        avg_train_loss = epoch_train_loss / batch_count
        train_losses.append(avg_train_loss)
        
        # Validation phase
        model.eval()
        epoch_val_loss = 0
        correct_preds = 0
        total_preds = 0
        val_batch_count = 0
        
        with torch.no_grad():
            for batch in val_loader:
                if len(batch) == 4:
                    states, mine_labels, _, _ = batch
                else:
                    states, mine_labels = batch
                
                states = states.to(device)
                mine_labels = mine_labels.to(device)
                
                predictions = model(states)
                loss = criterion(predictions, mine_labels)
                epoch_val_loss += loss.item()
                
                # Calculate accuracy on valid cells
                mask = (mine_labels != -1)
                pred_binary = (predictions.squeeze(1) > 0.5).float()
                target_binary = torch.clamp(mine_labels, 0, 1)
                
                correct = ((pred_binary == target_binary) & mask).sum().item()
                total = mask.sum().item()
                
                correct_preds += correct
                total_preds += total
                val_batch_count += 1
        
        avg_val_loss = epoch_val_loss / val_batch_count
        val_losses.append(avg_val_loss)
        
        val_accuracy = correct_preds / total_preds if total_preds > 0 else 0
        val_accuracies.append(val_accuracy)
        
        scheduler.step(avg_val_loss)
        
        print(f"Epoch {epoch+1}: Train Loss = {avg_train_loss:.4f}, "
              f"Val Loss = {avg_val_loss:.4f}, Val Acc = {val_accuracy:.3f}")
    
    return {
        'train_losses': train_losses,
        'val_losses': val_losses,
        'val_accuracies': val_accuracies,
        'model': model
    }

def train_actor_critic(actor_critic, train_loader, val_loader, num_epochs=15, lr=0.0005):
    actor_critic = actor_critic.to(device)
    
    # Use logic bot's survival predictions as target for critic
    # Use logic bot's move preferences as target for actor
    critic_criterion = SurvivalLoss()
    actor_criterion = nn.KLDivLoss(reduction='batchmean')
    
    optimizer = optim.Adam(actor_critic.parameters(), lr=lr)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=2)
    
    critic_train_losses = []
    actor_train_losses = []
    critic_val_losses = []
    
    print("Training Actor-Critic Network...")
    
    for epoch in range(num_epochs):
        # Training phase
        actor_critic.train()
        epoch_critic_loss = 0
        epoch_actor_loss = 0
        batch_count = 0
        
        for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Train"):
            if len(batch) != 4:
                continue  # Skip if we don't have survival data
            
            states, _, survival_targets, move_targets = batch
            states = states.to(device)
            survival_targets = survival_targets.to(device)
            move_targets = move_targets.to(device)
            
            optimizer.zero_grad()
            
            # Forward pass
            critic_pred, actor_probs = actor_critic(states)
            
            # Critic loss: predict survival duration
            critic_loss = critic_criterion(critic_pred, survival_targets)
            
            # Actor loss: match logic bot's move preferences
            # Normalize move targets to probability distribution
            move_targets_flat = move_targets.view(move_targets.size(0), -1)
            move_targets_norm = F.softmax(move_targets_flat / 0.1, dim=-1)  # Temperature scaling
            move_targets_norm = move_targets_norm.view_as(move_targets)
            
            actor_probs_flat = actor_probs.view(actor_probs.size(0), -1)
            actor_loss = actor_criterion(actor_probs_flat.log(), move_targets_norm.view(actor_probs.size(0), -1))
            
            # Combined loss
            total_loss = critic_loss + 0.5 * actor_loss
            
            total_loss.backward()
            torch.nn.utils.clip_grad_norm_(actor_critic.parameters(), 1.0)
            optimizer.step()
            
            epoch_critic_loss += critic_loss.item()
            epoch_actor_loss += actor_loss.item()
            batch_count += 1
        
        if batch_count > 0:
            avg_critic_loss = epoch_critic_loss / batch_count
            avg_actor_loss = epoch_actor_loss / batch_count
            critic_train_losses.append(avg_critic_loss)
            actor_train_losses.append(avg_actor_loss)
        
        # Validation phase
        actor_critic.eval()
        val_critic_loss = 0
        val_batch_count = 0
        
        with torch.no_grad():
            for batch in val_loader:
                if len(batch) != 4:
                    continue
                
                states, _, survival_targets, _ = batch
                states = states.to(device)
                survival_targets = survival_targets.to(device)
                
                critic_pred, _ = actor_critic(states)
                critic_loss = critic_criterion(critic_pred, survival_targets)
                
                val_critic_loss += critic_loss.item()
                val_batch_count += 1
        
        if val_batch_count > 0:
            avg_val_critic_loss = val_critic_loss / val_batch_count
            critic_val_losses.append(avg_val_critic_loss)
            
            scheduler.step(avg_val_critic_loss)
            
            print(f"Epoch {epoch+1}: Critic Loss = {avg_critic_loss:.4f}, "
                  f"Actor Loss = {avg_actor_loss:.4f}, Val Critic Loss = {avg_val_critic_loss:.4f}")
    
    return {
        'critic_train_losses': critic_train_losses,
        'actor_train_losses': actor_train_losses,
        'critic_val_losses': critic_val_losses,
        'model': actor_critic
    }

def train_sequential_thinker(model, train_loader, val_loader, num_epochs=15, lr=0.001):
    model = model.to(device)
    criterion = WeightedBCELoss(pos_weight=2.0, neg_weight=1.0)
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=2)
    
    train_losses = []
    val_losses = []
    step_losses = {i: [] for i in range(model.num_steps)}
    
    print("Training Sequential Thinker...")
    
    for epoch in range(num_epochs):
        # Training
        model.train()
        epoch_loss = 0
        batch_count = 0
        
        for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Train"):
            if len(batch) >= 2:
                states, mine_labels = batch[:2]
            else:
                continue
            
            states = states.to(device)
            mine_labels = mine_labels.to(device)
            
            optimizer.zero_grad()
            
            # Get predictions from all steps
            all_predictions, _ = model(states, return_all=True)
            
            # Calculate loss for each step, weighted by step number
            loss = 0
            for step, pred in enumerate(all_predictions):
                step_loss = criterion(pred, mine_labels)
                weight = (step + 1) / len(all_predictions)  # Later steps weighted more
                loss += weight * step_loss
            
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            
            epoch_loss += loss.item()
            batch_count += 1
        
        avg_train_loss = epoch_loss / batch_count if batch_count > 0 else 0
        train_losses.append(avg_train_loss)
        
        # Validation
        model.eval()
        val_loss = 0
        step_val_losses = {i: 0 for i in range(model.num_steps)}
        val_batch_count = 0
        
        with torch.no_grad():
            for batch in val_loader:
                if len(batch) >= 2:
                    states, mine_labels = batch[:2]
                else:
                    continue
                
                states = states.to(device)
                mine_labels = mine_labels.to(device)
                
                all_predictions, _ = model(states, return_all=True)
                
                batch_loss = 0
                for step, pred in enumerate(all_predictions):
                    step_loss = criterion(pred, mine_labels).item()
                    step_val_losses[step] += step_loss
                    weight = (step + 1) / len(all_predictions)
                    batch_loss += weight * step_loss
                
                val_loss += batch_loss
                val_batch_count += 1
        
        if val_batch_count > 0:
            avg_val_loss = val_loss / val_batch_count
            val_losses.append(avg_val_loss)
            
            for step in step_val_losses:
                step_val_losses[step] /= val_batch_count
                step_losses[step].append(step_val_losses[step])
            
            scheduler.step(avg_val_loss)
            
            print(f"Epoch {epoch+1}: Train Loss = {avg_train_loss:.4f}, Val Loss = {avg_val_loss:.4f}")
    
    return {
        'train_losses': train_losses,
        'val_losses': val_losses,
        'step_losses': step_losses,
        'model': model
    }

# ============================================================================
# BOT IMPLEMENTATIONS
# ============================================================================

class NeuralBot:
    def __init__(self, model, game, strategy='highest_safe', temperature=0.1):
        self.model = model
        self.game = game
        self.strategy = strategy
        self.temperature = temperature
        self.model.eval()
    
    def predict_mine_probs(self):
        with torch.no_grad():
            state = self.game.get_state()
            state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)
            
            if isinstance(self.model, SequentialThinker):
                probs = self.model(state_tensor, steps=5).squeeze().cpu().numpy()
            elif isinstance(self.model, ActorCriticNetwork):
                _, move_probs = self.model(state_tensor)
                probs = move_probs.squeeze().cpu().numpy()
                # Convert move probabilities to safety scores
                probs = 1 - probs  # Higher move prob = safer
            else:
                probs = self.model(state_tensor).squeeze().cpu().numpy()
        
        probs[self.game.revealed] = -1
        return probs
    
    def choose_move(self):
        probs = self.predict_mine_probs()
        unrevealed_mask = ~self.game.revealed & ~self.game.flagged
        
        if not np.any(unrevealed_mask):
            return None
        
        if self.strategy == 'highest_safe':
            # Choose cell with lowest mine probability
            safe_probs = probs.copy()
            safe_probs[~unrevealed_mask] = 2.0  # Very high for revealed/flagged
            r, c = np.unravel_index(np.argmin(safe_probs), safe_probs.shape)
            return r, c
        
        elif self.strategy == 'softmax':
            # Sample based on safety probabilities
            safe_probs = probs[unrevealed_mask]
            safe_cells = np.argwhere(unrevealed_mask)
            
            safety_scores = 1 - safe_probs  # Convert to safety scores
            safety_scores = np.maximum(safety_scores, 0.01)  # Avoid zero
            
            scores = np.exp(safety_scores / self.temperature)
            probs_dist = scores / np.sum(scores)
            
            idx = np.random.choice(len(safe_cells), p=probs_dist)
            return tuple(safe_cells[idx])
        
        elif self.strategy == 'epsilon_greedy' and hasattr(self, 'epsilon'):
            # Epsilon-greedy for exploration
            if random.random() < self.epsilon:
                # Random exploration
                safe_cells = np.argwhere(unrevealed_mask)
                idx = np.random.randint(len(safe_cells))
                return tuple(safe_cells[idx])
            else:
                # Exploitation: choose safest
                safe_probs = probs.copy()
                safe_probs[~unrevealed_mask] = 2.0
                r, c = np.unravel_index(np.argmin(safe_probs), safe_probs.shape)
                return r, c
        
        else:
            # Fallback: random move
            safe_cells = np.argwhere(unrevealed_mask)
            idx = np.random.randint(len(safe_cells))
            return tuple(safe_cells[idx])
    
    def run_game(self, max_steps=1000):
        steps = 0
        mines_triggered = 0
        
        while not self.game.game_over and steps < max_steps:
            move = self.choose_move()
            if move is None:
                break
            
            r, c = move
            result = self.game.reveal(r, c)
            
            if result == -1:
                mines_triggered += 1
            
            steps += 1
        
        return {
            'won': self.game.won,
            'steps': steps,
            'mines_triggered': mines_triggered,
            'cells_revealed': np.sum(self.game.revealed),
            'total_cells': self.game.height * self.game.width - self.game.num_mines
        }

class ActorCriticBot:
    def __init__(self, actor_critic_model, game, exploration_rate=0.1):
        self.model = actor_critic_model
        self.game = game
        self.exploration_rate = exploration_rate
        self.model.eval()
    
    def choose_move(self):
        with torch.no_grad():
            state = self.game.get_state()
            state_tensor = torch.FloatTensor(state).unsqueeze(0).to(device)
            
            # Get move probabilities from actor
            _, move_probs = self.model(state_tensor)
            move_probs = move_probs.squeeze().cpu().numpy()
            
            # Mask out revealed and flagged cells
            move_probs[self.game.revealed] = 0
            move_probs[self.game.flagged] = 0
            
            # Exploration: sometimes choose random move
            if random.random() < self.exploration_rate:
                unrevealed = np.argwhere(~self.game.revealed & ~self.game.flagged)
                if len(unrevealed) > 0:
                    idx = np.random.randint(len(unrevealed))
                    return tuple(unrevealed[idx])
            
            # Exploitation: choose cell with highest probability
            if np.sum(move_probs) > 0:
                move_probs = move_probs / np.sum(move_probs)  # Renormalize
                flat_idx = np.random.choice(move_probs.size, p=move_probs.flatten())
                r, c = np.unravel_index(flat_idx, move_probs.shape)
                return r, c
            else:
                # Fallback
                unrevealed = np.argwhere(~self.game.revealed & ~self.game.flagged)
                if len(unrevealed) > 0:
                    idx = np.random.randint(len(unrevealed))
                    return tuple(unrevealed[idx])
        
        return None
    
    def run_game(self, max_steps=1000):
        steps = 0
        mines_triggered = 0
        
        while not self.game.game_over and steps < max_steps:
            move = self.choose_move()
            if move is None:
                break
            
            r, c = move
            result = self.game.reveal(r, c)
            
            if result == -1:
                mines_triggered += 1
            
            steps += 1
        
        return {
            'won': self.game.won,
            'steps': steps,
            'mines_triggered': mines_triggered,
            'cells_revealed': np.sum(self.game.revealed)
        }

# ============================================================================
# EVALUATION AND COMPARISON FUNCTIONS
# ============================================================================

def evaluate_bot_performance(bot_class, model, difficulty='medium', num_trials=100, board_size=22):
    if difficulty == 'easy':
        num_mines = 50
    elif difficulty == 'medium':
        num_mines = 80
    elif difficulty == 'hard':
        num_mines = 100
    else:
        num_mines = 80
    
    results = {
        'wins': 0,
        'steps': [],
        'mines_triggered': [],
        'cells_revealed': [],
        'win_rates': [],
        'completion_rate': []
    }
    
    print(f"Evaluating {bot_class.__name__} on {difficulty} difficulty ({num_trials} trials)...")
    
    for trial in tqdm(range(num_trials)):
        game = MinesweeperGame(height=board_size, width=board_size, num_mines=num_mines)
        
        if bot_class == LogicBot:
            bot = LogicBot(game)
            result = bot.run_game()
        elif bot_class == NeuralBot:
            bot = NeuralBot(model, game, strategy='highest_safe')
            result = bot.run_game()
        elif bot_class == ActorCriticBot:
            bot = ActorCriticBot(model, game)
            result = bot.run_game()
        else:
            raise ValueError(f"Unknown bot class: {bot_class}")
        
        results['wins'] += int(result['won'])
        results['steps'].append(result['steps'])
        results['mines_triggered'].append(result.get('mines_triggered', 0))
        results['cells_revealed'].append(result.get('cells_revealed', 0))
        
        # Calculate intermediate win rates for learning curve
        if trial > 0:
            results['win_rates'].append(results['wins'] / (trial + 1))
        
        # Completion rate (percentage of safe cells revealed)
        if 'total_cells' in result:
            completion = result['cells_revealed'] / result['total_cells']
            results['completion_rate'].append(completion)
    
    # Calculate statistics
    results['win_rate'] = results['wins'] / num_trials
    results['avg_steps'] = np.mean(results['steps'])
    results['std_steps'] = np.std(results['steps'])
    results['avg_mines'] = np.mean(results['mines_triggered'])
    results['std_mines'] = np.std(results['mines_triggered'])
    
    if results['cells_revealed']:
        results['avg_cells'] = np.mean(results['cells_revealed'])
        results['std_cells'] = np.std(results['cells_revealed'])
    
    if results['completion_rate']:
        results['avg_completion'] = np.mean(results['completion_rate'])
    
    # 95% confidence intervals
    if num_trials > 1:
        results['ci_win_rate'] = 1.96 * np.sqrt(results['win_rate'] * (1 - results['win_rate']) / num_trials)
        results['ci_steps'] = 1.96 * results['std_steps'] / np.sqrt(num_trials)
        results['ci_mines'] = 1.96 * results['std_mines'] / np.sqrt(num_trials)
    
    return results

def compare_all_bots(difficulties=['easy', 'medium', 'hard'], num_trials=50):
    comparison_results = {}
    
    for difficulty in difficulties:
        print(f"\n{'='*60}")
        print(f"COMPARING BOTS ON {difficulty.upper()} DIFFICULTY")
        print(f"{'='*60}")
        
        # Train models for this difficulty
        print(f"\nTraining models for {difficulty}...")
        
        # Create dataset
        if difficulty == 'easy':
            num_mines = 50
        elif difficulty == 'medium':
            num_mines = 80
        else:
            num_mines = 100
        
        dataset = MinesweeperDataset(num_samples=2000, num_mines=num_mines, 
                                    difficulty=difficulty, include_logic_data=True)
        
        # Split dataset
        train_size = int(0.8 * len(dataset))
        val_size = len(dataset) - train_size
        train_dataset, val_dataset = torch.utils.data.random_split(
            dataset, [train_size, val_size]
        )
        
        train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
        val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
        
        # Train Mine Predictor
        print("\n1. Training Mine Predictor...")
        mine_predictor = MinePredictor(input_channels=5, hidden_channels=128)
        mine_results = train_mine_predictor(
            mine_predictor, train_loader, val_loader, num_epochs=10, lr=0.001
        )
        
        # Train Actor-Critic
        print("\n2. Training Actor-Critic...")
        actor_critic = ActorCriticNetwork(input_channels=5, hidden_channels=128)
        ac_results = train_actor_critic(
            actor_critic, train_loader, val_loader, num_epochs=8, lr=0.0005
        )
        
        # Train Sequential Thinker
        print("\n3. Training Sequential Thinker...")
        seq_thinker = SequentialThinker(input_channels=5, hidden_channels=128, num_steps=10)
        seq_results = train_sequential_thinker(
            seq_thinker, train_loader, val_loader, num_epochs=8, lr=0.001
        )
        
        # Evaluate all bots
        print("\n4. Evaluating bots...")
        
        # Logic Bot
        logic_results = evaluate_bot_performance(
            LogicBot, None, difficulty, num_trials=num_trials
        )
        
        # Neural Bot (Mine Predictor)
        neural_results = evaluate_bot_performance(
            NeuralBot, mine_results['model'], difficulty, num_trials=num_trials
        )
        
        # Actor-Critic Bot
        actor_critic_results = evaluate_bot_performance(
            ActorCriticBot, ac_results['model'], difficulty, num_trials=num_trials
        )
        
        # Sequential Thinker Bot
        seq_results_eval = evaluate_bot_performance(
            NeuralBot, seq_results['model'], difficulty, num_trials=num_trials
        )
        
        # Store results
        comparison_results[difficulty] = {
            'logic': logic_results,
            'neural': neural_results,
            'actor_critic': actor_critic_results,
            'sequential': seq_results_eval,
            'training_curves': {
                'mine_predictor': mine_results,
                'actor_critic': ac_results,
                'sequential': seq_results
            }
        }
        
        # Print comparison table
        print(f"\n{'='*60}")
        print(f"RESULTS FOR {difficulty.upper()} DIFFICULTY")
        print(f"{'='*60}")
        print(f"{'Bot Type':<20} {'Win Rate':<12} {'Avg Steps':<12} {'Avg Mines':<12} {'Completion':<12}")
        print(f"{'-'*60}")
        
        bots = [
            ('Logic Bot', logic_results),
            ('Neural Bot', neural_results),
            ('Actor-Critic', actor_critic_results),
            ('Sequential Thinker', seq_results_eval)
        ]
        
        for name, results in bots:
            win_rate_str = f"{results['win_rate']:.3f} ± {results.get('ci_win_rate', 0):.3f}"
            steps_str = f"{results['avg_steps']:.1f} ± {results.get('ci_steps', 0):.1f}"
            mines_str = f"{results['avg_mines']:.2f} ± {results.get('ci_mines', 0):.2f}"
            completion = results.get('avg_completion', 0)
            completion_str = f"{completion:.3f}" if completion > 0 else "N/A"
            
            print(f"{name:<20} {win_rate_str:<12} {steps_str:<12} {mines_str:<12} {completion_str:<12}")
    
    return comparison_results

# ============================================================================
# VISUALIZATION FUNCTIONS
# ============================================================================

def plot_training_curves(results_dict):
    """Plot training curves for all models"""
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    
    for idx, difficulty in enumerate(['easy', 'medium', 'hard']):
        if difficulty not in results_dict:
            continue
        
        curves = results_dict[difficulty]['training_curves']
        
        # Mine Predictor
        ax = axes[0, idx]
        if 'mine_predictor' in curves:
            mp_curves = curves['mine_predictor']
            ax.plot(mp_curves['train_losses'], label='Train Loss')
            ax.plot(mp_curves['val_losses'], label='Val Loss')
            ax.set_title(f'Mine Predictor - {difficulty.capitalize()}')
            ax.set_xlabel('Epoch')
            ax.set_ylabel('Loss')
            ax.legend()
            ax.grid(True, alpha=0.3)
        
        # Actor-Critic
        ax = axes[1, idx]
        if 'actor_critic' in curves:
            ac_curves = curves['actor_critic']
            ax.plot(ac_curves['critic_train_losses'], label='Critic Train Loss')
            ax.plot(ac_curves['critic_val_losses'], label='Critic Val Loss', linestyle='--')
            ax.set_title(f'Actor-Critic - {difficulty.capitalize()}')
            ax.set_xlabel('Epoch')
            ax.set_ylabel('Loss')
            ax.legend()
            ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Sequential Thinker Step Losses
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    for idx, difficulty in enumerate(['easy', 'medium', 'hard']):
        if difficulty not in results_dict:
            continue
        
        curves = results_dict[difficulty]['training_curves']
        if 'sequential' in curves:
            seq_curves = curves['sequential']
            ax = axes[idx]
            
            for step in range(min(5, len(seq_curves['step_losses']))):
                if step in seq_curves['step_losses']:
                    ax.plot(seq_curves['step_losses'][step], label=f'Step {step+1}')
            
            ax.set_title(f'Sequential Thinker - {difficulty.capitalize()}')
            ax.set_xlabel('Epoch')
            ax.set_ylabel('Validation Loss')
            ax.legend()
            ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def plot_performance_comparison(results_dict):
    """Create bar plots comparing bot performance"""
    difficulties = ['easy', 'medium', 'hard']
    bot_types = ['logic', 'neural', 'actor_critic', 'sequential']
    
    # Prepare data
    win_rates = {bot: [] for bot in bot_types}
    avg_steps = {bot: [] for bot in bot_types}
    avg_mines = {bot: [] for bot in bot_types}
    
    for difficulty in difficulties:
        if difficulty in results_dict:
            for bot in bot_types:
                if bot in results_dict[difficulty]:
                    win_rates[bot].append(results_dict[difficulty][bot]['win_rate'])
                    avg_steps[bot].append(results_dict[difficulty][bot]['avg_steps'])
                    avg_mines[bot].append(results_dict[difficulty][bot]['avg_mines'])
    
    # Create plots
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    x = np.arange(len(difficulties))
    width = 0.2
    
    # Win Rates
    ax = axes[0]
    for i, bot in enumerate(bot_types):
        if win_rates[bot]:
            ax.bar(x + (i-1.5)*width, win_rates[bot], width, label=bot.replace('_', ' ').title())
    
    ax.set_xlabel('Difficulty')
    ax.set_ylabel('Win Rate')
    ax.set_title('Win Rate Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels([d.capitalize() for d in difficulties])
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # Average Steps
    ax = axes[1]
    for i, bot in enumerate(bot_types):
        if avg_steps[bot]:
            ax.bar(x + (i-1.5)*width, avg_steps[bot], width, label=bot.replace('_', ' ').title())
    
    ax.set_xlabel('Difficulty')
    ax.set_ylabel('Average Steps')
    ax.set_title('Average Steps Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels([d.capitalize() for d in difficulties])
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # Average Mines Triggered
    ax = axes[2]
    for i, bot in enumerate(bot_types):
        if avg_mines[bot]:
            ax.bar(x + (i-1.5)*width, avg_mines[bot], width, label=bot.replace('_', ' ').title())
    
    ax.set_xlabel('Difficulty')
    ax.set_ylabel('Average Mines Triggered')
    ax.set_title('Mines Triggered Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels([d.capitalize() for d in difficulties])
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def visualize_thinking_process(model, game_state, num_steps=5):
    """Visualize how predictions evolve with thinking time"""
    model.eval()
    
    with torch.no_grad():
        state_tensor = torch.FloatTensor(game_state).unsqueeze(0).to(device)
        
        if isinstance(model, SequentialThinker):
            all_predictions, all_features = model(state_tensor, steps=num_steps, return_all=True)
        else:
            # For non-sequential models, simulate by adding noise/iterations
            all_predictions = []
            for step in range(num_steps):
                pred = model(state_tensor)
                all_predictions.append(pred)
    
    # Create visualization
    fig, axes = plt.subplots(2, num_steps, figsize=(15, 6))
    
    for step in range(num_steps):
        pred = all_predictions[step].squeeze().cpu().numpy()
        
        # Top row: prediction heatmap
        ax = axes[0, step]
        im = ax.imshow(pred, cmap='hot', vmin=0, vmax=1)
        ax.set_title(f'Step {step+1}')
        ax.axis('off')
        
        if step == num_steps - 1:
            plt.colorbar(im, ax=ax)
        
        # Bottom row: change from previous step
        ax = axes[1, step]
        if step == 0:
            im = ax.imshow(pred, cmap='hot', vmin=0, vmax=1)
            ax.set_title('Initial')
        else:
            prev_pred = all_predictions[step-1].squeeze().cpu().numpy()
            change = np.abs(pred - prev_pred)
            im = ax.imshow(change, cmap='coolwarm', vmin=0, vmax=0.3)
            ax.set_title(f'Δ from step {step}')
        
        ax.axis('off')
    
    plt.suptitle('Evolution of Predictions During Thinking', fontsize=16)
    plt.tight_layout()
    plt.show()
    
    # Plot convergence
    convergence = []
    for i in range(1, num_steps):
        pred_i = all_predictions[i].squeeze().cpu().numpy()
        pred_prev = all_predictions[i-1].squeeze().cpu().numpy()
        convergence.append(np.mean(np.abs(pred_i - pred_prev)))
    
    plt.figure(figsize=(8, 5))
    plt.plot(range(1, num_steps), convergence, 'o-', linewidth=2, markersize=8)
    plt.xlabel('Thinking Step')
    plt.ylabel('Mean Absolute Change')
    plt.title('Prediction Convergence During Thinking')
    plt.grid(True, alpha=0.3)
    plt.show()

# ============================================================================
# STATISTICAL ANALYSIS
# ============================================================================

def perform_statistical_analysis(results_dict):
    """Perform statistical tests to compare bot performance"""
    
    print("\n" + "="*60)
    print("STATISTICAL ANALYSIS")
    print("="*60)
    
    for difficulty in ['easy', 'medium', 'hard']:
        if difficulty not in results_dict:
            continue
        
        print(f"\n{difficulty.upper()} DIFFICULTY:")
        print("-"*40)
        
        # Get win rates for statistical tests
        bots_data = {}
        for bot_type in ['logic', 'neural', 'actor_critic', 'sequential']:
            if bot_type in results_dict[difficulty]:
                # Simulate binary win/loss data from win rate
                win_rate = results_dict[difficulty][bot_type]['win_rate']
                # For simplicity, we'll compare win rates directly
                bots_data[bot_type] = win_rate
        
        # Compare each neural bot to logic bot
        logic_rate = bots_data.get('logic', 0)
        
        for bot_type in ['neural', 'actor_critic', 'sequential']:
            if bot_type in bots_data:
                neural_rate = bots_data[bot_type]
                improvement = neural_rate - logic_rate
                
                # Z-test for proportion difference
                n = 50  # Assuming 50 trials
                p_pool = (logic_rate * n + neural_rate * n) / (2 * n)
                se = np.sqrt(p_pool * (1 - p_pool) * (2 / n))
                z = improvement / se if se > 0 else 0
                p_value = 2 * (1 - stats.norm.cdf(abs(z)))
                
                print(f"\n{bot_type.replace('_', ' ').title()} vs Logic Bot:")
                print(f"  Logic Win Rate: {logic_rate:.3f}")
                print(f"  {bot_type.replace('_', ' ').title()} Win Rate: {neural_rate:.3f}")
                print(f"  Improvement: {improvement:.3f}")
                print(f"  p-value: {p_value:.4f}")
                
                if p_value < 0.05:
                    print(f"  → Statistically significant improvement!")
                else:
                    print(f"  → No statistically significant difference")
        
        # Compare neural bots among themselves
        print(f"\nComparison among neural bots:")
        neural_bots = [(name, rate) for name, rate in bots_data.items() if name != 'logic']
        neural_bots.sort(key=lambda x: x[1], reverse=True)
        
        for i, (name, rate) in enumerate(neural_bots):
            print(f"  {i+1}. {name.replace('_', ' ').title()}: {rate:.3f}")

# ============================================================================
# CASE STUDY ANALYSIS
# ============================================================================

def analyze_case_studies(model, num_cases=5):
    """Analyze specific cases where bots make different decisions"""
    
    print("\n" + "="*60)
    print("CASE STUDY ANALYSIS")
    print("="*60)
    
    for case_num in range(num_cases):
        print(f"\nCase Study {case_num + 1}:")
        print("-"*40)
        
        # Create a game
        game = MinesweeperGame(height=10, width=10, num_mines=15)
        game.reset(first_click=(5, 5))
        game.reveal(5, 5)
        
        # Play some moves to create interesting situation
        bot = LogicBot(game)
        for _ in range(10):
            bot.make_move()
        
        # Get predictions from different approaches
        neural_bot = NeuralBot(model, game)
        mine_probs = neural_bot.predict_mine_probs()
        
        # Get logic bot's next move
        logic_bot = LogicBot(copy.deepcopy(game))
        if logic_bot.inferred_safe:
            logic_move = logic_bot.inferred_safe.pop()
        else:
            logic_move = random.choice(list(logic_bot.cells_remaining))
        
        # Get neural bot's next move
        neural_move = neural_bot.choose_move()
        
        print(f"Board state: {np.sum(game.revealed)} cells revealed")
        print(f"Logic bot would choose: {logic_move}")
        print(f"Neural bot chooses: {neural_move}")
        
        if logic_move != neural_move:
            print(f"  → Different decisions!")
            
            # Analyze why
            logic_r, logic_c = logic_move
            neural_r, neural_c = neural_move
            
            logic_prob = mine_probs[logic_r, logic_c] if logic_r < mine_probs.shape[0] and logic_c < mine_probs.shape[1] else -1
            neural_prob = mine_probs[neural_r, neural_c]
            
            print(f"  Logic move mine probability: {logic_prob:.3f}")
            print(f"  Neural move mine probability: {neural_prob:.3f}")
            
            if neural_prob < logic_prob:
                print(f"  → Neural bot chooses safer cell")
            else:
                print(f"  → Logic bot chooses safer cell")
        
        # Visualize the board
        fig, axes = plt.subplots(1, 3, figsize=(12, 4))
        
        # Revealed board
        revealed_display = np.full_like(game.clues, -1, dtype=float)
        revealed_display[game.revealed] = game.clues[game.revealed]
        
        axes[0].imshow(revealed_display, cmap='viridis', vmin=-1, vmax=8)
        axes[0].set_title('Revealed Board')
        if logic_move:
            axes[0].plot(logic_move[1], logic_move[0], 'ro', markersize=10, label='Logic Move')
        if neural_move:
            axes[0].plot(neural_move[1], neural_move[0], 'gx', markersize=10, label='Neural Move')
        axes[0].legend()
        
        # Mine probabilities
        im = axes[1].imshow(mine_probs, cmap='hot', vmin=0, vmax=1)
        axes[1].set_title('Mine Probabilities')
        plt.colorbar(im, ax=axes[1])
        
        # Difference
        unrevealed = ~game.revealed
        if np.any(unrevealed):
            safest = np.unravel_index(np.argmin(mine_probs * unrevealed + (1 - unrevealed) * 2), mine_probs.shape)
            axes[2].imshow(mine_probs * unrevealed, cmap='Greens')
            axes[2].set_title('Safest Cells')
            axes[2].plot(safest[1], safest[0], 'b*', markersize=15, label='Safest')
            axes[2].legend()
        
        plt.tight_layout()
        plt.show()

# ============================================================================
# MAIN EXECUTION
# ============================================================================

def main():
    print("="*60)
    print("MINESWEEPER AI - FINAL PROJECT IMPLEMENTATION")
    print("="*60)
    
    # Step 1: Quick test of basic functionality
    print("\n1. Testing basic game mechanics...")
    test_game = MinesweeperGame(height=8, width=8, num_mines=10)
    test_game.reset(first_click=(4, 4))
    test_game.reveal(4, 4)
    print(f"   Game created: {test_game.height}x{test_game.width}, {test_game.num_mines} mines")
    print(f"   First click clue: {test_game.clues[4, 4]}")
    
    # Step 2: Test Logic Bot
    print("\n2. Testing Logic Bot...")
    logic_wins = 0
    for i in range(10):
        game = MinesweeperGame(height=8, width=8, num_mines=10)
        bot = LogicBot(game)
        result = bot.run_game()
        logic_wins += int(result['won'])
    print(f"   Logic Bot win rate in 10 games: {logic_wins/10:.1%}")
    
    # Step 3: Train and evaluate on all difficulties
    print("\n3. Training and evaluating all models...")
    
    # For demonstration, we'll run with fewer trials to save time
    # In the actual submission, use more trials for better statistics
    comparison_results = compare_all_bots(
        difficulties=['easy', 'medium', 'hard'],
        num_trials=30  # Reduced for demonstration
    )
    
    # Step 4: Visualize results
    print("\n4. Visualizing results...")
    plot_training_curves(comparison_results)
    plot_performance_comparison(comparison_results)
    
    # Step 5: Statistical analysis
    perform_statistical_analysis(comparison_results)
    
    # Step 6: Case studies
    print("\n5. Analyzing case studies...")
    
    # Load a trained model for case studies
    if 'medium' in comparison_results:
        # Use the sequential thinker model for case studies
        seq_model = comparison_results['medium']['training_curves']['sequential']['model']
        analyze_case_studies(seq_model, num_cases=3)
    
    # Step 7: Demonstrate thinking process
    print("\n6. Demonstrating thinking process...")
    
    # Create a test game
    demo_game = MinesweeperGame(height=12, width=12, num_mines=20)
    demo_game.reset(first_click=(6, 6))
    demo_game.reveal(6, 6)
    
    # Reveal some cells around it
    for dr in [-1, 0, 1]:
        for dc in [-1, 0, 1]:
            if dr == 0 and dc == 0:
                continue
            r, c = 6 + dr, 6 + dc
            if 0 <= r < 12 and 0 <= c < 12 and np.random.random() > 0.5:
                demo_game.reveal(r, c)
    
    # Visualize thinking process
    if 'medium' in comparison_results:
        seq_model = comparison_results['medium']['training_curves']['sequential']['model']
        visualize_thinking_process(seq_model, demo_game.get_state(), num_steps=5)
    
    # Step 8: Generate final report summary
    print("\n" + "="*60)
    print("FINAL SUMMARY")
    print("="*60)
    
    print("\nKey Findings:")
    print("-"*40)
    
    for difficulty in ['easy', 'medium', 'hard']:
        if difficulty in comparison_results:
            print(f"\n{difficulty.upper()} DIFFICULTY:")
            
            best_bot = None
            best_rate = 0
            
            for bot_type, results in comparison_results[difficulty].items():
                if bot_type != 'training_curves' and 'win_rate' in results:
                    win_rate = results['win_rate']
                    print(f"  {bot_type.replace('_', ' ').title()}: {win_rate:.3f}")
                    
                    if win_rate > best_rate:
                        best_rate = win_rate
                        best_bot = bot_type
            
            if best_bot:
                print(f"  → Best performer: {best_bot.replace('_', ' ').title()} ({best_rate:.3f})")
    
    print("\n" + "="*60)
    print("PROJECT COMPLETED SUCCESSFULLY!")
    print("="*60)
    
    return comparison_results

# ============================================================================
# RUN THE PROJECT
# ============================================================================

if __name__ == "__main__":
    try:
        # Set smaller values for testing, larger for final run
        print("Note: Running in demonstration mode with reduced trials.")
        print("For final submission, increase num_trials to 100+ per difficulty.")
        
        final_results = main()
        
        # Save results
        import pickle
        with open('minesweeper_results.pkl', 'wb') as f:
            pickle.dump(final_results, f)
        print("\nResults saved to 'minesweeper_results.pkl'")
        
    except Exception as e:
        print(f"\nError during execution: {e}")
        import traceback
        traceback.print_exc()
```
